1. DateTime  วันที่ผลิต  
2. Formula เบอร์อาหาร(สูตรอาหารแต่ล่ะชนิด)
3. Performance อัตราการ Feed อาหารเข้าในเครื่องอัดเม็ด หน่วยเป็น Ton/hr.
4. steam อัตตรา การจ่ายไอน้ำเข้าไปคลุกเค้าอาหาร หน่วยเป็น kg/hr.
5. Temp_Con อุณหภูมิของอาหารหลังจากคลุกเค้ากับไอน้ำ หน่วยเป็น องศา C
6. Temp_Front_Pellet อุณหภูมิของอาหารหลังจากถูกบีบอัดขึ้นรูปเป็นอาหารเม็ด หน่วยเป็น องศา C
7. current ค่ากระแสไฟฟ้าของ Motor ที่ใช้ขับเครื่องอัดเม็ด หน่วยเป็น Amp
โดยที่ต้อง ทำนายว่า การปรับ parameter ข้อ 3 - 6 ของแต่ละเบอร์อาหารที่ต่างๆกัน จะทำให้ เกิด current (ข้อ7) มากกว่า 80 % fullload กระแสมอเตอร์ ใช่หรือไม่
