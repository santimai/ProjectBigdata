```
# ลบข้อมูลแถวที่ไม่เกี่ยวกับการนำไปสร้าง Model 
df = data.loc[:, ['DateTime', 'Formula','Performance','steam','Temp_Con','Temp_Front_Pellet','current']]
df.head(5)
```

```
df.info()
```

```
# ค้นหาข้อมูลที่เป็น NA 
df.isna().sum(axis=0)
```

```
# ค้นหาข้อมูลที่เป็น Outlier
df.describe(include=['object', 'float', 'int']).T
```

พิจารณาข้อมูลขณะเดินเครื่อง โดยที่ ค่า 
Preformance มาก 1 และไม่เกิน 35 

1.   Preformance มาก 1 และไม่เกิน 35 
2.   steam มากกว่า 0
3. Temp_Con ไม่เกิน 90 องศา
4. Temp_Front_Pellet  ไม่เกิน 95 องศา
5. current มากว่า 0 



```
median=df.median() #คำนวณให้เฉพาะคอลัมน์ที่คิดได้
modes=df.mode() #คิดให้ทุกคอลัมน์
# เอามารวมกัน
modesNmedian=modes
modesNmedian[median.index.values.tolist()]=median.to_list()
```

```
replace_val=modesNmedian.loc[0]
replace_val
```

```
df_filtered = df.query('Performance > 1')
df_filtered = df_filtered.query('Performance <  35')
df_filtered = df_filtered.query('current >1 ')
df_filtered = df_filtered.query('steam >1 ')
df_filtered.describe(include=['object', 'float', 'int']).T
```

```
#แทนค่า Outlier ด้วยค่า replace_val ของ Temp_Con และ Temp_Front_Pellet
df_filtered.loc[df_filtered.Temp_Con > 90, 'Temp_Con'] = replace_val.Temp_Con
df_filtered.loc[df_filtered.Temp_Front_Pellet > 95, 'Temp_Front_Pellet'] = replace_val.Temp_Front_Pellet
df_filtered.describe(include=['object', 'float', 'int']).T
```

```
#แปลง categorical Formula   attribute  ให้เป็น   numerical attribute 
df_filtered['Fac_For'] = df_filtered['Formula'].astype('category').cat.codes
df_filtered.head(5)
```

```
df_filtered.info()
```

```
#สร้าง Class Target เพื่อทำนาย คำนวนจาก กระแสเครื่องอัดเม็ด Electric current over 80% of full load  366*0.8,292 > True ,292 < 0
def f(row):
    if row['current'] > 292:
        val = True
    elif row['current'] < 366:
        val = False
    else:
        val = False
    return val
df_filtered['Status'] = df_filtered.apply(f, axis=1)
df_filtered.head(5)
```
